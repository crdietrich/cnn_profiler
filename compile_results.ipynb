```python
import pandas as pd
import parse
```

```python
power_files, profile_files = parse.collate()
```

```python
df_stats = parse.pstats_compile(profile_files)
```

```python
df_stats.dtypes
```

```python
df_power = parse.power_compile(power_files)
```

```python
df_power.dtypes
```

```python
df = df_stats.merge(right=df_power, how='outer',
         on=['platform', 'PU', 'PU_type', 'run_id', 'TF_version'])
```

```python
_df = df.drop(['filepath', 'TF_version', 'Watts_min', 'Watts_max', 'Watts_mean_off'], axis=1)
_df.sort_values(by=["platform", "PU", "run_id"], axis=0, ascending=True, inplace=True)
_df.reset_index(inplace=True, drop=True)
_df
```
